```python
import random as rnd
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import stats 
import seaborn as sns 
import statistics
```

```python
# parametros esteticos de seaborn
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (8, 4)})
```

```python
numeros = []
rnd.seed(1272022)
[numeros.append(rnd.random()) for i in range(1000)]
print(numeros)
```

```python
sorted_random_data = np.sort(numeros)
p = 1. * np.arange(len(sorted_random_data)) / float(len(sorted_random_data) - 1)
media = np.mean(p)
desviacion = statistics.pstdev(p)
print("The CDF result is-",p)
fig = plt.figure()
fig.suptitle('CDF of data points')
ax2 = fig.add_subplot(111)
ax2.plot(sorted_random_data, p)
ax2.set_xlabel('sorted_random_data')
ax2.set_ylabel('p')
```

```python
# Graficando Bernoulli
bernoulli = stats.bernoulli(0.5)
x = np.arange(-1, 3)
fmp = bernoulli.pmf(x) # Función de Masa de Probabilidad
fig, ax = plt.subplots()
ax.plot(x, fmp, 'bo')
ax.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
ax.set_yticks([0., 0.2, 0.4, 0.6, 0.8])
plt.title('Distribución Bernoulli')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
# Graficando Normal
mu, sigma = media, desviacion # media y desvio estandar

normal = stats.norm(mu, sigma)
x = np.linspace(normal.ppf(0.01),
                normal.ppf(0.99), 100)
fp = normal.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Normal')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()

# histograma de distribución normal.
datos = np.random.normal(mu, sigma, 1000)
plt.hist(datos, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma')
plt.show()
```

```python
# Graficando Binomial
N, p = 100, 0.5 # parametros de forma 
binomial = stats.binom(N, p) # Distribución
x = np.arange(binomial.ppf(0.01),
              binomial.ppf(0.99))
fmp = binomial.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Binomial')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```
